```python
import yfinance as yf
from datetime import timedelta

# Funcin robusta para rellenar valores nulos usando yfinance
def rellenar_missings_yfinance(df, symbol_col='symbol', date_col='date'):
    import pandas as pd
    df = df.copy()
    symbols = df[symbol_col].unique()
    for symbol in symbols:
        sub = df[df[symbol_col] == symbol]
        for col in ['open', 'high', 'low', 'close', 'volume']:
            if col in sub.columns and sub[col].isna().any():
                fechas_nulas = sub.loc[sub[col].isna(), date_col]
                if not fechas_nulas.empty:
                    start = fechas_nulas.min()
                    end = fechas_nulas.max()
                    # Descargar datos de yfinance para el periodo faltante
                    try:
                        data_yf = yf.download(symbol, start=start, end=(end + timedelta(days=1)).strftime('%Y-%m-%d'))
                    except Exception as e:
                        print(f'Error descargando {symbol}: {e}')
                        continue
                    if data_yf is None or data_yf.empty:
                        continue
                    # Convertir el ndice a datetime.date para comparar correctamente
                    data_yf.index = pd.to_datetime(data_yf.index).date
                    for fecha in fechas_nulas:
                        # Convertir fecha a date si es Timestamp
                        fecha_dt = fecha.date() if hasattr(fecha, 'date') else fecha
                        if fecha_dt in data_yf.index:
                            col_yf = col.capitalize() if col != 'volume' else 'Volume'
                            if col_yf in data_yf.columns:
                                valor = data_yf.loc[fecha_dt, col_yf]
                                # Solo rellenar si sigue siendo nulo
                                mask = (df[symbol_col]==symbol) & (df[date_col]==fecha)
                                if pd.isna(df.loc[mask, col]).any() and valor is not None:
                                    df.loc[mask, col] = valor
    return df

# Usar la funcin para rellenar los missings
df_p = rellenar_missings_yfinance(df_primeros)
df_p.isna().sum()
```
