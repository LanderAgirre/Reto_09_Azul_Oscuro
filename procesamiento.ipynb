```python
import yfinance as yf
# Filtramos las filas con valores faltantes en "close"
faltantes = dias_laborales[dias_laborales["close"].isna()].copy()

# Diccionario para cachear descargas y no repetir llamadas innecesarias
data_cache = {}

# Recorremos símbolos únicos de los faltantes
for symbol in faltantes["symbol"].unique():
    fechas = faltantes.loc[faltantes["symbol"] == symbol, "fecha"].dt.strftime("%Y-%m-%d").tolist()
    
    # Descargamos solo el rango de fechas que necesitamos
    start, end = min(fechas), max(fechas)
    
    if symbol not in data_cache:
        try:
            hist = yf.download(symbol, start=start, end=end)
            hist.index = hist.index.tz_localize(None)  # quitar timezone
            data_cache[symbol] = hist
        except Exception as e:
            print(f"Error descargando {symbol}: {e}")
            continue

    # Asignamos el valor a cada fila
    for f in fechas:
        try:
            valor = data_cache[symbol].loc[f, "Close"]
            dias_laborales.loc[(dias_laborales["symbol"] == symbol) & (dias_laborales["fecha"] == pd.to_datetime(f)), "close"] = valor
        except KeyError:
            pass  # si no hay datos de esa fecha (ej. fin de semana), se queda NaN
```

```python
import yfinance as yf
import pandas as pd

faltantes = dias_laborales[dias_laborales["close"].isna()].copy()
data_cache = {}

for symbol in faltantes["symbol"].unique():
    fechas = faltantes.loc[faltantes["symbol"] == symbol, "fecha"].dt.strftime("%Y-%m-%d").tolist()
    start, end = min(fechas), (pd.to_datetime(max(fechas)) + pd.Timedelta(days=1)).strftime("%Y-%m-%d")
    
    if symbol not in data_cache:
        try:
            hist = yf.download(symbol, start=start, end=end)
            hist.index = hist.index.tz_localize(None)
            data_cache[symbol] = hist
        except Exception as e:
            print(f"Error descargando {symbol}: {e}")
            continue

    for f in fechas:
        fecha_ts = pd.to_datetime(f)
        # Buscamos la fila más cercana hacia atrás si no existe
        valor = data_cache[symbol]["Close"].reindex([fecha_ts], method="ffill").iloc[0]
        dias_laborales.loc[(dias_laborales["symbol"] == symbol) & 
                           (dias_laborales["fecha"] == fecha_ts), "close"] = valor
```

```python
import yfinance as yf
from datetime import timedelta
import pandas as pd

def rellenar_close_yfinance(df, symbol_col='symbol', date_col='date', close_col='close'):
    df = df.copy()
    # Asegurarse de que la columna de fecha esté en formato datetime
    df[date_col] = pd.to_datetime(df[date_col])
    for idx, row in df.iterrows():
        if pd.isna(row[close_col]):
            symbol = row[symbol_col]
            fecha = row[date_col]
            # Descargar solo el día necesario
            try:
                start = fecha.strftime('%Y-%m-%d')
                end = (fecha + timedelta(days=1)).strftime('%Y-%m-%d')
                data_yf = yf.download(symbol, start=start, end=end)
            except Exception as e:
                print(f'Error descargando {symbol} en {fecha}: {e}')
                continue
            if not data_yf.empty and 'Close' in data_yf.columns:
                # Buscar el valor de cierre para la fecha
                fecha_dt = fecha.date()
                data_yf.index = pd.to_datetime(data_yf.index).date
                if fecha_dt in data_yf.index:
                    valor = data_yf.loc[fecha_dt, 'Close']
                    if not pd.isna(valor):
                        df.at[idx, close_col] = valor
    return df

# Ejemplo de uso:
df_close = rellenar_close_yfinance(df_primeros, symbol_col='symbol', date_col='fecha', close_col='close')
df_close
```

```python
import yfinance as yf
from datetime import timedelta

# Función robusta para rellenar valores nulos usando yfinance
def rellenar_missings_yfinance(df, symbol_col='symbol', date_col='date'):
    import pandas as pd
    df = df.copy()
    symbols = df[symbol_col].unique()
    for symbol in symbols:
        sub = df[df[symbol_col] == symbol]
        for col in ['open', 'high', 'low', 'close', 'volume']:
            if col in sub.columns and sub[col].isna().any():
                fechas_nulas = sub.loc[sub[col].isna(), date_col]
                if not fechas_nulas.empty:
                    start = fechas_nulas.min()
                    end = fechas_nulas.max()
                    # Descargar datos de yfinance para el periodo faltante
                    try:
                        data_yf = yf.download(symbol, start=start, end=(end + timedelta(days=1)).strftime('%Y-%m-%d'))
                    except Exception as e:
                        print(f'Error descargando {symbol}: {e}')
                        continue
                    if data_yf is None or data_yf.empty:
                        continue
                    # Convertir el índice a datetime.date para comparar correctamente
                    data_yf.index = pd.to_datetime(data_yf.index).date
                    for fecha in fechas_nulas:
                        # Convertir fecha a date si es Timestamp
                        fecha_dt = fecha.date() if hasattr(fecha, 'date') else fecha
                        if fecha_dt in data_yf.index:
                            col_yf = col.capitalize() if col != 'volume' else 'Volume'
                            if col_yf in data_yf.columns:
                                valor = data_yf.loc[fecha_dt, col_yf]
                                # Solo rellenar si sigue siendo nulo
                                mask = (df[symbol_col]==symbol) & (df[date_col]==fecha)
                                if pd.isna(df.loc[mask, col]).any() and valor is not None:
                                    df.loc[mask, col] = valor
    return df

# Usar la función para rellenar los missings
df_p = rellenar_missings_yfinance(df_primeros)
df_p.isna().sum()
```
