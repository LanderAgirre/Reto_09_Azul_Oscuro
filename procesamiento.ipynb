```python
import yfinance as yf
from datetime import timedelta

# Función para obtener los periodos faltantes de cada symbol y rellenar con yfinance
def rellenar_missings_yfinance(df, symbol_col='symbol', date_col='date'):
    df = df.copy()
    symbols = df[symbol_col].unique()
    for symbol in symbols:
        sub = df[df[symbol_col] == symbol]
        for col in sub.columns:
            if sub[col].isna().any():
                fechas_nulas = sub.loc[sub[col].isna(), date_col]
                if not fechas_nulas.empty:
                    start = fechas_nulas.min()
                    end = fechas_nulas.max()
                    # Descargar datos de yfinance para el periodo faltante
                    data_yf = yf.download(symbol, start=start, end=(end + timedelta(days=1)).strftime('%Y-%m-%d'))
                    # Rellenar los valores nulos con los datos descargados
                    for fecha in fechas_nulas:
                        if fecha.strftime('%Y-%m-%d') in data_yf.index.strftime('%Y-%m-%d'):
                            for colname in ['open','high','low','close','volume']:
                                if colname in df.columns and pd.isna(df.loc[(df[symbol_col]==symbol) & (df[date_col]==fecha), colname]).any():
                                    valor = data_yf.loc[fecha.strftime('%Y-%m-%d'), colname.capitalize()] if colname.capitalize() in data_yf.columns else None
                                    if valor is not None:
                                        df.loc[(df[symbol_col]==symbol) & (df[date_col]==fecha), colname] = valor
    return df

# Usar la función para rellenar los missings
datos = rellenar_missings_yfinance(datos)
datos.isna().sum()
```
